### 🎯 <span style="color:#2980b9">1. Tích vô hướng (Dot Product) & Hồi quy đa biến</span>
* 🏠 Thay vì một biến, các vấn đề thực tế có nhiều biến đầu vào (vd: số phòng ngủ $x_1$, tuổi nhà $x_2$).
* 🧮 **Dự đoán** được tính bằng công thức toán học: $Prediction = w_1 \cdot x_1 + w_2 \cdot x_2$
* 📉 **Gradient Descent** được áp dụng để tính đạo hàm riêng (*Partial Derivatives*) và cập nhật từng trọng số một.

### ⚡ <span style="color:#27ae60">2. Tính phi tuyến (Non-Linearity) & Hàm ReLU</span>
* 📏 **Vấn đề của Tuyến tính:** Nếu chỉ dùng các phép nhân cộng, AI chỉ tạo ra các đường thẳng. Ghép 100 đường thẳng lại vẫn ra đường thẳng. AI sẽ không thể học các mẫu phức tạp.
* 🔄 **Giải pháp:** Hàm kích hoạt (*Activation Function*) như **ReLU**: $f(x) = \max(0, x)$. Việc "bẻ gãy" các giá trị âm thành 0 giúp AI học được các đường cong và dữ liệu phức tạp.
* ⚠️ **Luật bất thành văn (Dying ReLU):** Tuyệt đối **KHÔNG** dùng ReLU ở tầng Output cho các bài toán dự đoán số (Regression). Nếu đầu ra bị âm, ReLU biến nó thành 0, đạo hàm bằng 0, và AI sẽ "chết não" (không cập nhật nữa)!

### 🥞 <span style="color:#f39c12">3. Mạng Nơ-ron Đa Tầng (Multi-Layer Neural Network)</span>
* ➡️ Dữ liệu đi từ **Input Layer** ➔ **Hidden Layers** (*Bắt buộc dùng ReLU*) ➔ **Output Layer** (*Dạng tuyến tính*).
* 🔗 Kết quả **ĐÃ KÍCH HOẠT** của tầng ẩn sẽ trở thành **ĐẦU VÀO** mới cho tầng tiếp theo.

### 🧠 <span style="color:#d35400">4. Tại sao lại cần PyTorch & Tensors?</span>
* 💻 **Hardware:** Tensors giúp tính toán hàng triệu phép nhân cùng lúc trên hàng ngàn lõi của **GPU**, thay vì chạy vòng lặp `for` chậm chạp trên CPU.
* ⚙️ **Autograd:** Hàm `loss.backward()` tự động tính đạo hàm cho hàng tỷ tham số, giúp trò không bao giờ phải tính đạo hàm bằng tay nữa!

### 🔄 <span style="color:#8e44ad">5. Vòng lặp huấn luyện chuẩn (PyTorch Training Loop)</span>
1. 🧹 `optimizer.zero_grad()`: Xóa sạch đạo hàm cũ.
2. ⏩ `prediction = model(x)`: Forward Pass.
3. 🎯 `loss = criterion(...)`: Tính sai số.
4. ⏪ `loss.backward()`: Tính đạo hàm tự động.
5. 🚀 `optimizer.step()`: Cập nhật trọng số.



```python
import torch
import torch.nn as nn

# 1. DỮ LIỆU ĐẦU VÀO (House Pricing)
x = torch.tensor([3.0, 10.0])
y_true = torch.tensor([500.0])

# 2. XÂY DỰNG MÔ HÌNH (Linear Regression)
model = nn.Linear(in_features=2, out_features=1, bias=False)

# Khởi tạo trọng số ngẫu nhiên ban đầu để giống bài toán Pure Python
with torch.no_grad():
    model.weight.copy_(torch.tensor([[1.0, 1.0]]))

# 3. CÔNG CỤ TỰ ĐỘNG CỦA PYTORCH
criterion = nn.MSELoss()  
# This is the Loss Function. Its only job is to calculate exactly how "wrong" the model is. 
# MSELoss: Mean Squared Error <==> (prediction, y_true)**2

optimizer = torch.optim.SGD(model.parameters(), lr=0.005) 
# This is the Optimization Algorithm. Its job is to actually change the weights to make the error smaller
# In PyTorch: SGD (Stochastic Gradient Descent) automates this.

print("🚀 BẮT ĐẦU HUẤN LUYỆN VỚI PYTORCH AUTOGRAD...")

for step in range(30):
    optimizer.zero_grad() # xóa đạo hàm trước đó
    prediction = model(x) 
    loss = criterion(prediction, y_true) 
    loss.backward() # backward
    optimizer.step() # update weight
    
    if step % 5 == 0:
        weights = model.weight.data.numpy()[0]
        print(f"Vòng {step:2d} | w1={weights[0]:.2f}, w2={weights[1]:.2f} | Dự đoán: {prediction.item():.1f} | Sai số: {loss.item():.1f}")

print(f"\n✅ KẾT THÚC! AI dự đoán giá nhà: {model(x).item():.1f}")
```
