# Bài 20: Word Embeddings - Tạo không gian ngữ nghĩa
# Mỗi từ vựng (ID) sẽ được gán cho một vector (danh sách các số thực).
# Bảng chứa tất cả các vector này gọi là Embedding Matrix (Ma trận nhúng).

```python
import random

def init_embedding_matrix(vocab_size, embedding_dim):
    """
    Khởi tạo ma trận Embedding (Bảng tra cứu vector).
    Input: 
      vocab_size: Số lượng từ trong từ điển (V)
      embedding_dim: Số chiều của vector nhúng (D)
    Output: Một list của list kích thước (V x D).
    """
    matrix = []
    
    # ========================================================
    # NHIỆM VỤ 1: TỰ CODE KHỞI TẠO MA TRẬN
    # 1. Dùng 1 vòng lặp để tạo ra `vocab_size` hàng.
    # 2. Ở mỗi hàng, tạo ra một list gồm `embedding_dim` con số ngẫu nhiên.
    #    Gợi ý: Dùng random.uniform(-0.5, 0.5) để lấy số ngẫu nhiên nhỏ.
    # 3. Đẩy cái list đó vào `matrix`.
    # ========================================================
    
    for i in range(vocab_size):
        rows = []
        for j in range(embedding_dim):
            rows.append(random.uniform(-0.5, 0.5))
        matrix.append(rows)
    
    return matrix
```

```python
def get_embeddings(sentence_ids, embedding_matrix, embedding_dim):
    """
    Tra cứu vector cho từng từ trong câu dựa vào ID.
    Input:
      sentence_ids: list các ID của câu (VD: [0, 2, -1, 3])
      embedding_matrix: Bảng tra cứu tạo ở hàm trên
      embedding_dim: Số chiều (D)
    Output: Một list chứa các vector tương ứng.
    """
    sentence_embeddings = []
    
    # ========================================================
    # NHIỆM VỤ 2: TỰ CODE TRA CỨU VECTOR
    # Duyệt qua từng ID trong sentence_ids:
    # 1. Nếu ID hợp lệ (lớn hơn hoặc bằng 0 VÀ nhỏ hơn kích thước của embedding_matrix):
    #    Lấy hàng thứ ID trong embedding_matrix đẩy vào sentence_embeddings.
    # 2. Nếu ID KHÔNG hợp lệ (từ mới <UNK>, ID = -1):
    #    Tạo ra một vector chứa toàn số 0.0 (độ dài = embedding_dim) và đẩy vào.
    # ========================================================
    for ID in sentence_ids:
        if (ID >= 0) and (ID < len(embedding_matrix)):
            sentence_embeddings.append(embedding_matrix[ID])
        else:
            sentence_embeddings.append([0.0]*embedding_dim)
            
    return sentence_embeddings
```

KHU VỰC TEST

```python

print("--- BÀI 20: TRA CỨU VECTOR NHÚNG TỪ VỰNG ---")
    
# Giả lập lại kết quả của Bài 19
vocab = {'sư': 0, 'phụ': 1, 'dạy': 2, 'code': 3}
vocab_size = len(vocab)
embedding_dim = 3 # Chọn không gian 3 chiều cho dễ nhìn
    
print("1. Đang khởi tạo Ma trận Embedding...")
random.seed(42) # Cố định seed
emb_matrix = init_embedding_matrix(vocab_size, embedding_dim)
    
for word, idx in vocab.items():
    try:
        print(f"Từ '{word}' (ID {idx}) -> Vector: {[round(v, 4) for v in emb_matrix[idx]]}")
    except Exception as e:
        print(f"Lỗi khởi tạo ma trận: {e}")
        break
            
print("\n2. Đang tra cứu cho một câu cụ thể...")
# Câu: "sư phụ dạy toán" -> "toán" là từ lạ (ID -1)
test_ids = [0, 1, 2, -1] 
print(f"IDs đầu vào: {test_ids}")
    
try:
    embeddings = get_embeddings(test_ids, emb_matrix, embedding_dim)
    print("\nMa trận đầu vào cho mạng nơ-ron (Câu đã hóa vector):")
    for i, emb in enumerate(embeddings):
        print(f"Từ thứ {i+1} (ID {test_ids[i]}): {[round(v, 4) for v in emb]}")
            
    if embeddings[3] == [0.0, 0.0, 0.0]:
        print("\n✅ XUẤT SẮC! Đã xử lý đúng từ <UNK> thành vector [0.0, 0.0, 0.0]")
except Exception as e:
    print(f"Lỗi tra cứu vector: {e}")

```
