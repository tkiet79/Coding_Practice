# Bài 19: Tokenization - Biến chữ thành số
# Mạng nơ-ron không biết đọc chữ, nó chỉ biết tính toán với các con số. 
# Vì vậy, bước đầu tiên của mọi hệ thống AI (kể cả ChatGPT) là phải 
# tạo ra một "Từ điển" (Vocabulary) để ép các từ vựng thành các ID định danh.

```python
def build_vocabulary(corpus):
    """
    Hàm xây dựng từ điển từ một kho văn bản (corpus).
    Input: corpus (một list chứa các câu văn string)
    Output: vocab (một dictionary ánh xạ: 'từ' -> số_thứ_tự)
    Ví dụ: {'sư': 0, 'phụ': 1, 'dạy': 2, 'code': 3}
    """
    vocab = {}
    current_index = 0
    # ========================================================
    # NHIỆM VỤ 1: TỰ CODE XÂY DỰNG TỪ ĐIỂN
    # 1. Duyệt qua từng câu trong corpus.
    # 2. Tách câu thành các từ rời rạc (Dùng hàm .split() của string).
    # 3. Duyệt qua từng từ: Nếu từ đó CHƯA CÓ trong vocab, hãy thêm nó vào 
    #    với giá trị là current_index, sau đó tăng current_index lên 1.
    # ========================================================
    for cau in corpus:
        cac_tu = cau.split()
        for tu in cac_tu:
            if tu not in vocab:
                vocab[tu] = current_index
                current_index += 1
                
    return vocab
```

```python
def encode_sentence(sentence, vocab):
    """
    Hàm biến một câu văn thành mảng các con số dựa vào từ điển.
    Input: sentence (một câu văn string), vocab (dictionary đã tạo ở trên)
    Output: Một list các con số.
    """
    encoded = []
    
    # ========================================================
    # NHIỆM VỤ 2: TỰ CODE MÃ HÓA CÂU
    # 1. Tách câu input thành các từ.
    # 2. Duyệt qua từng từ. Kiểm tra:
    #    - Nếu từ có trong vocab, lấy số thứ tự của nó đẩy (append) vào mảng encoded.
    #    - Nếu từ KHÔNG CÓ trong vocab (từ mới toanh), đẩy giá trị -1 vào mảng.
    #      (Giá trị -1 này tượng trưng cho token <UNK> - Unknown Word).
    # ========================================================
    cac_tu = sentence.split()
    for tu in cac_tu:
        if tu in vocab:
            encoded.append(vocab[tu])
        else:
            encoded.append(-1)
    
    return encoded
```

```python
corpus = [
    "sư phụ dạy code",
    "em học code rất vui",
    "sư phụ rất nghiêm khắc"
]

print("1. Đang xây dựng Từ điển (Vocabulary)...")
try:
    vocab = build_vocabulary(corpus)
    print("Từ điển của em:", vocab)
    print(f"Tổng số từ vựng: {len(vocab)}")
except Exception as e:
    print(f"Lỗi Nhiệm vụ 1: {e}")
    
print("\n2. Đang mã hóa câu văn mới...")
try:
    test_sentence_1 = "em học code"
    test_sentence_2 = "sư phụ dạy toán" # Từ 'toán' chưa từng xuất hiện!
    
    print(f"Câu 1 '{test_sentence_1}' -> {encode_sentence(test_sentence_1, vocab)}")
    print(f"Câu 2 '{test_sentence_2}' -> {encode_sentence(test_sentence_2, vocab)}")
except Exception as e:
    print(f"Lỗi Nhiệm vụ 2: {e}")
```
