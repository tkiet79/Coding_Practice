 ### <font color="orange">🧠 GIAI ĐOẠN 2: XỬ LÝ NGÔN NGỮ TỰ NHIÊN (NLP)</font>
 #### Bài 24: Tự Code Lõi Toán Học của LSTM

 Lõi của LSTM nhận vào 3 thứ: `x_t` (từ mới), `h_{t-1}` (trí nhớ ngắn hạn cũ), `C_{t-1}` (băng chuyền dài hạn cũ).
 Và nó trải qua 6 bước tính toán:

 1. **Cổng Quên (Forget Gate)**: Quyết định vứt bỏ ký ức nào trên băng chuyền.
    > $ \color{orange} f_t = \sigma(W_{fx} \cdot x_t + W_{fh} \cdot h_{t-1} + b_f) $

 2. **Cổng Nạp (Input Gate)**: Quyết định lấy thông tin mới nào.
    > $ \color{orange} i_t = \sigma(W_{ix} \cdot x_t + W_{ih} \cdot h_{t-1} + b_i) $

 3. **Ký ức Ứng viên (Candidate Cell)**: Chắt lọc thông tin từ `x_t` và `h_{t-1}`.
    > $ \color{orange} \tilde{C}_t = \tanh(W_{cx} \cdot x_t + W_{ch} \cdot h_{t-1} + b_c) $

 4. **CẬP NHẬT BĂNG CHUYỀN (Cell State Update)** - BƯỚC QUAN TRỌNG NHẤT:
    Ký ức cũ * Cổng Quên + Ký ức mới * Cổng Nạp.
    > $ \color{orange} C_t = f_t * C_{t-1} + i_t * \tilde{C}_t $

 5. **Cổng Xuất (Output Gate)**: Quyết định xuất ra cái gì cho bước tiếp theo.
    > $ \color{orange} o_t = \sigma(W_{ox} \cdot x_t + W_{oh} \cdot h_{t-1} + b_o) $

 6. **Cập nhật Trí nhớ ngắn hạn (Hidden State Update)**:
    > $ \color{orange} h_t = o_t * \tanh(C_t) $

# %%

```python

import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def lstm_step(x_t, h_prev, C_prev, weights):
    """
    Mô phỏng 1 bước nhảy của LSTM.
    weights: Là một dictionary chứa tất cả 12 ma trận trọng số.
    """
    # Lấy các ma trận trọng số ra cho dễ nhìn
    W_fx, W_fh, b_f = weights['f']
    W_ix, W_ih, b_i = weights['i']
    W_cx, W_ch, b_c = weights['c']
    W_ox, W_oh, b_o = weights['o']

    # ========================================================
    # NHIỆM VỤ THỬ THÁCH NHẤT TRONG NGÀY: TỰ CODE 6 BƯỚC LSTM
    # Gợi ý: 
    # - Nhân ma trận dùng np.dot()
    # - Cộng bình thường dùng dấu +
    # - Nhân TỪNG PHẦN TỬ (element-wise) như f_t * C_{t-1} thì chỉ cần dùng dấu *
    # ========================================================

    # 1. Cổng Quên
    f_t = sigmoid(np.dot(W_fx, x_t) + np.dot(W_fh,h_prev) + b_f)
    
    # 2. Cổng Nạp
    i_t = sigmoid(np.dot(W_ix,x_t) + np.dot(W_ih, h_prev) + b_i)
    
    # 3. Ký ức Ứng viên
    C_tilde = np.tanh(np.dot(W_cx,x_t) + np.dot(W_ch, h_prev) + b_c)
    
    # 4. Cập nhật Băng chuyền dài hạn (C_t)
    C_t = f_t * C_prev + i_t * C_tilde
    
    # 5. Cổng Xuất
    o_t = sigmoid(np.dot(W_ox,x_t) + np.dot(W_oh, h_prev) + b_o)
    
    # 6. Cập nhật Trí nhớ ngắn hạn (h_t)
    h_t = o_t * np.tanh(C_t)

    return h_t, C_t

# %%
def main():
    print("--- BÀI 24: MỔ XẺ TOÁN HỌC LSTM ---")
    np.random.seed(42)
    
    emb_dim = 3
    hidden_size = 4
    
    # Khởi tạo 12 bộ trọng số ngẫu nhiên cho 4 trạm (Forget, Input, Candidate, Output)
    weights = {}
    for gate in ['f', 'i', 'c', 'o']:
        weights[gate] = (
            np.random.randn(hidden_size, emb_dim),      # W_x  (4 x 3)
            np.random.randn(hidden_size, hidden_size),  # W_h  (4 x 4)
            np.random.randn(hidden_size)                # b    (4,)
        )
    
    # Câu văn: ["sư", "phụ", "dạy"]
    sentence = [
        np.array([0.9, 0.1, 0.0]),
        np.array([0.8, 0.5, 0.1]),
        np.array([0.2, 0.8, 0.4])
    ]
    
    # Trí nhớ ban đầu (Ngắn hạn và Dài hạn đều rỗng)
    h_prev = np.zeros(hidden_size)
    C_prev = np.zeros(hidden_size)
    
    try:
        print("Bắt đầu đẩy dữ liệu lên băng chuyền LSTM...\n")
        for t, x_t in enumerate(sentence):
            h_prev, C_prev = lstm_step(x_t, h_prev, C_prev, weights)
            print(f"--- Sau khi đọc từ thứ {t+1} ---")
            print(f"Ngắn hạn (h_t): {[round(v, 4) for v in h_prev]}")
            print(f"Dài hạn (C_t): {[round(v, 4) for v in C_prev]}\n")
            
        print("✅ XUẤT SẮC! Đã chạy xong lõi Toán học của LSTM.")
    except Exception as e:
        print(f"❌ Lỗi code Toán: {e}")

if __name__ == "__main__":
    main()
```
