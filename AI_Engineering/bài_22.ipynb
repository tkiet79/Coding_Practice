Tóm tắt bài trước: Em đã biết cách biến một câu thành các vector bằng Word Embeddings (Ví dụ: "sư" -> $v_1$, "phụ" -> $v_2$, "dạy" -> $v_3$, "code" -> $v_4$).

Vấn đề cực lớn nảy sinh:
Nếu em đưa các vector này vào mạng Nơ-ron bình thường (như Giai đoạn 1), nó sẽ đối xử với các từ một cách độc lập.
Điều này dẫn đến thảm họa ngôn ngữ: Mạng nơ-ron sẽ thấy câu "Sư phụ dạy toán" và câu "Toán dạy sư phụ" là y chang nhau (vì chứa cùng lượng từ vựng). Ngôn ngữ loài người quan trọng nhất là THỨ TỰ (Sequence) và NGỮ CẢNH (Context). Mạng nơ-ron truyền thống bị "mất trí nhớ", đọc chữ sau quên luôn chữ trước.

Giải pháp đột phá: Recurrent Neural Network (RNN - Mạng nơ-ron hồi tiếp)
Các pháp sư AI đã thiết kế lại mạng nơ-ron sao cho nó có một "Bộ nhớ ngắn hạn" (Hidden State - $h_t$).

Cách nó đọc sách giống hệt con người:
Đọc từ thứ 1 ("sư"): Tạo ra bộ nhớ $h_1$.
Đọc từ thứ 2 ("phụ"): Lấy bộ nhớ $h_1$ kết hợp với từ "phụ" để suy ra ngữ cảnh, tạo ra bộ nhớ mới $h_2$.
Cứ thế cuộn dần đến cuối câu.

Công thức Toán học cốt lõi của một ô RNN (tại bước thời gian $t$) vô cùng thanh lịch:
            $$\color{orange} h_t = \tanh(W_{xh} \cdot x_t + W_{hh} \cdot h_{t-1} + b)$$

Trong đó:

- $x_t$: Vector của từ đang đọc.
- $h_{t-1}$: Bộ nhớ từ chữ trước đó truyền sang.
- $\tanh$: Hàm kích hoạt (ép giá trị về -1 đến 1 để bộ nhớ không bị phình to phát nổ).
- $W_{xh}$, $W_{hh}$: Là các ma trận trọng số

```python

import numpy as np

def rnn_step(x_t, h_prev, W_xh, W_hh, b_h):
    """
    Hàm tính toán cho MỘT BƯỚC thời gian (một từ) trong RNN.
    Input:
      x_t: Vector embedding của TỪ HIỆN TẠI (1D array)
      h_prev: Bộ nhớ từ TỪ TRƯỚC ĐÓ (1D array)
      W_xh, W_hh, b_h: Các ma trận trọng số
    Output:
      h_next: Bộ nhớ mới sau khi đã đọc xong từ hiện tại
    """
    # ========================================================
    # NHIỆM VỤ: TỰ CODE CÔNG THỨC LÕI CỦA RNN
    # Công thức: h_next = tanh(W_xh * x_t + W_hh * h_prev + b_h)
    # Gợi ý: Dùng np.dot() để nhân ma trận với vector. 
    # Nhớ cộng 3 thành phần lại rồi mới đưa qua hàm np.tanh() nhé!
    # ========================================================
    
    h_next = np.tanh(np.dot(W_xh,x_t) + np.dot(W_hh,h_prev) + b_h)  # Thay None bằng công thức của em
    
    return h_next

# %%
def main():
    print("--- BÀI 22: MÔ PHỎNG RNN ĐỌC CÂU VĂN ---")
    
    np.random.seed(42) # Cố định seed
    
    # Giả lập thông số mạng RNN
    emb_dim = 3    # Mỗi từ là vector 3 chiều
    hidden_size = 4 # Bộ nhớ có sức chứa 4 chiều
    
    # Khởi tạo trọng số ngẫu nhiên cho não bộ AI
    W_xh = np.random.randn(hidden_size, emb_dim)    
    W_hh = np.random.randn(hidden_size, hidden_size)
    b_h = np.zeros(hidden_size)
    
    # Giả lập câu văn có 3 từ đã được nhúng thành vector
    # Ví dụ: ["sư", "phụ", "dạy"]
    sentence = [
        np.array([0.9, 0.1, 0.0]), # x_1: Vector chữ "sư"
        np.array([0.8, 0.5, 0.1]), # x_2: Vector chữ "phụ"
        np.array([0.2, 0.8, 0.4])  # x_3: Vector chữ "dạy"
    ]
    
    # Bắt đầu đọc sách: Bộ nhớ ban đầu trống rỗng (toàn số 0)
    h_prev = np.zeros(hidden_size)
    
    print("🧠 Quá trình não bộ RNN nạp thông tin:\n")
    try:
        for t, x_t in enumerate(sentence):
            print(f"--- Đọc từ thứ {t+1} ---")
            print(f"Vector từ mới: {x_t}")
            print(f"Ký ức cũ: {[round(v, 4) for v in h_prev]}")
            
            # Cập nhật bộ nhớ
            h_prev = rnn_step(x_t, h_prev, W_xh, W_hh, b_h)
            
            print(f"-> Ký ức MỚI sau khi cập nhật: {[round(v, 4) for v in h_prev]}\n")
            
        print("✅ Đã đọc xong câu văn! Biến 'h_prev' cuối cùng chính là VECTOR NGỮ CẢNH tóm tắt toàn bộ ý nghĩa của cả câu.")
    except Exception as e:
        print(f"❌ Lỗi code rồi: {e}")

if __name__ == "__main__":
    main()
```
